```python
def table_to_igraph_links(enhancer_final, a):
    node_1 = []
    node_2 = []
    node_name = []
    for i in range(len(enhancer_final)):
        n2_s = enhancer_final[a].iloc[i].split(":")
        for j in range(len(n2_s)):
            if n2_s[j] != "-":
                node_2.append(n2_s[j])
                node_1.append(enhancer_final["Neiro"].iloc[i])
    table = pd.DataFrame({"Node1": node_2, "Node2": node_1}).drop_duplicates()
    return table
```

```python
def table_to_igraph_nodes(enhancer_final, a, b):
    nodes = []
    node_name = []
    x1 = []
    x1prop = []
    for i in range(len(enhancer_final)):
        n2_s = enhancer_final[a].iloc[i].split(":")
        names_s = enhancer_final[b].iloc[i].split(":")
        for j in range(len(n2_s)):
            if n2_s[j] != "-":
                if n2_s[j] not in nodes:
                    nodes.append(n2_s[j])
                    node_name.append(names_s[j])
                    x1facs = FACS_prop[FACS_prop["gene"] == n2_s[j]]
                    if len(x1facs) > 0:
                        x1.append(x1facs["X1"].iloc[0])
                        x1prop.append(x1facs["X1.prop"].iloc[0])
                    else:
                        x1.append(0)
                        x1prop.append(0)
                if enhancer_final["Neiro"].iloc[i] not in nodes:
                    nodes.append(enhancer_final["Neiro"].iloc[i])
                    node_name.append(enhancer_final["Symbol"].iloc[i])
                    x1.append(enhancer_final["X1"].iloc[i])
                    x1prop.append(enhancer_final["X1 percent"].iloc[i])
    table = pd.DataFrame({"Nodes": nodes, "Name": node_name, "X1": x1, "X1.prop": x1prop})
    return table
```

```python
nodes = table_to_igraph_nodes(enhancer_final, "Bound motifs", "Bound motif names")
len(nodes)
```

```python
links = table_to_igraph_links(enhancer_final, "Bound motifs")[(table_to_igraph_links(enhancer_final, "Bound motifs")["Node1"].isin(nodes["Nodes"]))&(table_to_igraph_links(enhancer_final, "Bound motifs")["Node2"].isin(nodes["Nodes"]))]
```

```python
links.to_csv("/hydra/TF_data/edges_grn_tf_bound2021.csv", index=False)
nodes.to_csv("/hydra/TF_data/vertices_grn_tf_bound2021.csv", index=False)
```
