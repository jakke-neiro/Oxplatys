```r
library(GenomicFeatures)
library(Gviz)
library(rtracklayer)
```

```r
x1_foot = import("/drives/raid/AboobakerLab/jakke/atac_sexual_final/footprinting/x1_footprints.bw")
options(ucscChromosomeNames=FALSE)
x1_foot_track = DataTrack(x1_foot, type = "l", name="X1 Footprints", fill.histogram = "navyblue", col.histogram= "navyblue")
```

```r
k27ac_mean = import("/drives/raid/AboobakerLab/jakke/chip_neiro/k27ac/X1-H3K27ac-meanlog2.bw")
k4me1_mean = import("/drives/raid/AboobakerLab/jakke/chip_neiro/k4me1/h3k4me1meanlog2.bw")
options(ucscChromosomeNames=FALSE)
k27ac_track = DataTrack(k27ac_mean, type = "l", name="H3K27ac", transformation = function(x) { x[x < 0] <- 0; x }, fill.histogram = "darkgreen", col.histogram= "darkgreen")
k4me1_track = DataTrack(k4me1_mean, type = "l", name="H3K4me1", transformation = function(x) { x[x < 0] <- 0; x }, fill.histogram = "darkgrey", col.histogram= "darkgrey")
```

```r
x1RNA = import("/drives/raid/AboobakerLab/jakke/FACS/x1_mean_final.bw")
x1RNA_track = DataTrack(x1RNA, type="l", name="RNA-seq", fill.histogram = "darkred", col.histogram= "darkred")
```

```r
txdb = makeTxDbFromGFF("/drives/raid/AboobakerLab/jakke/sexual_genome_annotation_files/ncrna_Neiro/stringtie.Bioconductor.gtf", format = "gtf")
```

## pou4-1

dd_Smes_g4_145: 1530353-1532343

```r
options(ucscChromosomeNames=FALSE)
coebound = import.bed("/drives/raid/AboobakerLab/jakke/atac_sexual_final/footprinting/motif_beds/coe.bed")
coebound = AnnotationTrack(coebound, name="coe", fill="black", rotation.title=1)
```

```r
options(ucscChromosomeNames=FALSE)
gtrack <- GenomeAxisTrack(cex=2.25)
txTr <- GeneRegionTrack(txdb, chromosome = "dd_Smes_g4_145", fill="orange", name=" ", fontsize=20)
ht <- HighlightTrack(trackList = list(k27ac_track, k4me1_track, x1_foot_track, coebound),start = 1530353-1050, width = 100,chromosome = "dd_Smes_g4_145")
options(repr.plot.width = 15, repr.plot.height = 15)
plotTracks(c(gtrack, ht , txTr, x1RNA_track), transcriptAnnotation = "symbol", shape = "arrow", chromosome="dd_Smes_g4_145", from=1530353-3000, to=1530353+4000, cex.group=1.75, type="hist", window = -1, windowSize = 1, 
                       cex.axis=1.5, cex.title=1.3)
```
