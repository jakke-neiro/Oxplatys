```python
enhancer_final["Bound motifs"] = "-"
enhancer_final["Bound motif names"] = "-"
enhancer_final["Bound counts"] = 0
enhancer_final["Unbound motifs"] = "-"
enhancer_final["Unbound motif names"] = "-"
enhancer_final["Unbound counts"] = 0
for i in range(len(enhancer_final)):
    chrom = enhancer_final["seqnames"].iloc[i]
    start = enhancer_final["start"].iloc[i]
    motif_table = X1_table_bound[(abs(X1_table_bound["TFBS_start"] - start) < 1000)&(X1_table_bound["TFBS_chr"] == chrom)]
    gene_strs = []
    name_strs = []
    if len(motif_table) > 0:
        gene_str = motif_table["GeneID"].iloc[0]
        gene_strs.append(motif_table["GeneID"].iloc[0])
        name_str = motif_table["TF name"].iloc[0]
        name_strs.append(motif_table["TF name"].iloc[0])
        enhancer_final["Bound counts"].iloc[i] = enhancer_final["Bound counts"].iloc[i]+1
        for j in range(1,len(motif_table)):
            if motif_table["GeneID"].iloc[j] not in gene_strs:
                gene_str = gene_str + ":" + motif_table["GeneID"].iloc[j]
                gene_strs.append(motif_table["GeneID"].iloc[j])
                enhancer_final["Bound counts"].iloc[i] = enhancer_final["Bound counts"].iloc[i]+1
            if motif_table["TF name"].iloc[j] not in name_strs:
                name_str = name_str + ":" + motif_table["TF name"].iloc[j]
                name_strs.append(motif_table["TF name"].iloc[j])
        enhancer_final["Bound motifs"].iloc[i] = gene_str
        enhancer_final["Bound motif names"].iloc[i] = name_str
    unbound_motif_table = X1_table[(abs(X1_table["TFBS_start"] - start) < 1000)&(X1_table["TFBS_chr"] == chrom)]
    if len(unbound_motif_table) > 0:
        gene_str = unbound_motif_table["GeneID"].iloc[0]
        gene_strs.append(unbound_motif_table["GeneID"].iloc[0])
        name_str = unbound_motif_table["TF name"].iloc[0]
        name_strs.append(unbound_motif_table["TF name"].iloc[0])
        enhancer_final["Unbound counts"].iloc[i] = enhancer_final["Unbound counts"].iloc[i]+1
        for j in range(1,len(unbound_motif_table)):
            if unbound_motif_table["GeneID"].iloc[j] not in gene_strs:
                gene_str = gene_str + ":" + unbound_motif_table["GeneID"].iloc[j]
                gene_strs.append(unbound_motif_table["GeneID"].iloc[j])
                enhancer_final["Unbound counts"].iloc[i] = enhancer_final["Unbound counts"].iloc[i]+1
            if unbound_motif_table["TF name"].iloc[j] not in name_strs:
                name_str = name_str + ":" + unbound_motif_table["TF name"].iloc[j]
                name_strs.append(unbound_motif_table["TF name"].iloc[j])
        enhancer_final["Unbound motifs"].iloc[i] = gene_str
        enhancer_final["Unbound motif names"].iloc[i] = name_str
```
