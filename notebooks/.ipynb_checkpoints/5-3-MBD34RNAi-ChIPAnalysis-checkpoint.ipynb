# 4. Enhancer annotation of mean H3K4me1/3 values

The mean H3K4me1/3 log2 fold-change values were added to the table of enhancers. 

```python
%%bash
cd /hydra/enhancer_regions
ls enhancer_annotation_july21.csv
```

```python
import pandas as pd
enhancer_annotation = pd.read_csv("/hydra/enhancer_regions/enhancer_annotation_july21.csv")
```

```python
enhancer_annotation_FACS = enhancer_annotation
enhancer_annotation_FACS["X1.prop"] = 0
enhancer_annotation_FACS["X1"] = 0
FACS_prop = pd.read_csv("/hydra/FACS/FACS_prop.csv")
```

```python
enhancer_annotation_FACS.shape
```

```python
for i in range(len(enhancer_annotation_FACS)):
    df = FACS_prop[FACS_prop["gene"] == enhancer_annotation_FACS.iloc[i,13]]
    if len(df) > 0:
        enhancer_annotation_FACS.iloc[i,16] = df.iloc[0,1]
        enhancer_annotation_FACS.iloc[i,17] = df.iloc[0,5]
```

```python
k4me1lptlog2aroundconsensus = pd.read_csv("/hydra/chip_RNAi/k4me1rnai_july21.tab", sep="\t", skiprows=3, header=None)
k4me3lptlog2aroundconsensus = pd.read_csv("/hydra/chip_sexual_organised/k4me3/k4me3rnai_july21.tab", sep="\t", skiprows=3, header=None)
```

```python
k4me1RNAiconsensus_mean = k4me1lptlog2aroundconsensus.mean(axis=1)
k4me3RNAiconsensus_mean = k4me3lptlog2aroundconsensus.mean(axis=1)
```

```python
chip_summits = pd.read_csv("/hydra/enhancer_regions/chip_summits_july21.bed", sep="\t", header=None)
```

```python
chip_summits["k4me1RNAi"] = k4me1RNAiconsensus_mean
chip_summits["k4me3RNAi"] = k4me3RNAiconsensus_mean
```

```python
enhancer_annotation_FACS["K4me1RNAi"] = 0
enhancer_annotation_FACS["K4me3RNAi"] = 0
for i in range(len(enhancer_annotation_FACS)):
    peak_name = enhancer_annotation_FACS.iloc[i, 5]
    df_chip = chip_summits[chip_summits.iloc[:,3] == peak_name]
    if (len(df_chip) > 0):
        enhancer_annotation_FACS.iloc[i, 18] = df_chip.iloc[0,5]
        enhancer_annotation_FACS.iloc[i, 19] = df_chip.iloc[0,6]
```

```python
enhancer_annotation_FACS.to_csv("/hydra/enhancer_regions/enhancer_annotation_FACSRNAi_july21.csv", index=False)
```
