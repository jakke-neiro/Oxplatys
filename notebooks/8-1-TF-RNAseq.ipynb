# 2. Cowles 2014
The data to this section have been collected from Cowles et al. (2014). Cowles et al. (2014) performed coe(RNAi). The Collier/Olfactory-1/Early B-cell factor (COE) family of transcription factors is necessary for animal development. Planarian *coe* is expressed in differentiating and mature neurons and that its function is required for proper CNS regeneration. 

RNA-seq samples were extracted from whole worms. 3 coe(RNAi) and 3 gfp(RNAi) worms. One week after the final dsRNA treatment, RNA was extracted from three independent control and coe(RNAi) animal groups using Trizol (Life Technologies).  

### References

Cowles, M. W., Omuro, K. C., Stanley, B. N., Quintanilla, C. G., & Zayas, R. M. (2014). COE loss-of-function analysis reveals a genetic program underlying maintenance and regeneration of the nervous system in planarians. PLoS Genetics, 10(10), e1004746.

coe(RNAi)
* SRR1133800 
* SRR2486925
* SRR2486926

control
* SRR2486927
* SRR2486928
* SRR2486929

# 5. Stowers 2016

First, Stowers Institute examined gene expression profiles of adult flatworms after 6, 12, or 18 days of treatment with RNAi against either Pax5, Sox8, Zfp, or an unc22 non-targeting control. The experiment was performed in quadruplicate resulting in 48 samples. The samples were collected from whole worms.

Second, Stowers Institute examined gene expression profiles of FACS sorted X1 stem cell population from adult flatworms after 6, 9, and 13 days of treatment with RNAi against either zfp1, p53, or an unc22 non-targeting control. The experiment was performed in quadruplicate, with one discarded as an outlier for low sequence yield, resulting in 35 samples

### References
RNA Seq analysis in flatworm following 6, 12, or 18 day treatment with RNAi against Pax5, Sox8, or Zfp1, relative to unc22 to identify transcriptional targets.
https://www.ncbi.nlm.nih.gov/bioproject/PRJNA319245

RNA Seq analysis of neoblasts from Schmidtea mediterranea treated with RNAi against zfp1, p53, or unc22 to identify factors involved in neoblast differentiation.
https://www.ncbi.nlm.nih.gov/bioproject/PRJNA319244

* pax5 (pax2/5/8-1) whole worm tissue 6, 12, 18 dpa
* sox8 (soxP-3) whole worm tissue 6, 12, 18 dpa
* zfp (zfp-1) whole worm tissue 6, 12, 18 dpa
* unc22 (control) whole worm tissue 6, 12, 18 dpa

* p53 X1 cells 6, 9, 13 dpa
* zfp1 X1 cells 6, 9, 13 dpa
* unc22 X1 cells 6, 9, 13 dpa

* zfp-1: no motif
