```python
import pandas as pd
X1_bound = pd.read_csv("/drives/raid/AboobakerLab/jakke/atac_sexual_final/footprinting/BINDetect_X1/X1_bound_distance.csv")
```

```python
bound_concat = pd.concat([X1_bound.reset_index(drop="TRUE"), X1_table_bound.reset_index(drop="TRUE")], axis=1)
```

```python
bound_concat = bound_concat[bound_concat["TF name"] != "a"]
```

```python
bound_concat.to_csv("/drives/raid/AboobakerLab/jakke/atac_sexual_final/footprinting/BINDetect_X1/X1_bound_distance_names.csv")
```

```python
import numpy as np
promoters = []
distal = []
#totals
motif_names = list(pd.Categorical(bound_concat["TF name"]).categories)
for i in motif_names:
    subset = bound_concat[bound_concat["TF name"] == i]
    promoters.append(len(subset[subset["distanceToTSS"] < 500])/len(subset))
    distal.append(len(subset[subset["distanceToTSS"] > 10000])/len(subset))
```

```python
pd.DataFrame({"name": motif_names, "fraction": promoters, "distal": distal}).to_csv("/drives/raid/AboobakerLab/jakke/atac_sexual_final/footprinting/BINDetect_X1/X1_bound_promoterfraction.csv")
```
